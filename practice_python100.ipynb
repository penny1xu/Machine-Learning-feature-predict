```python

h_guess_num = 5
e_guess_num = 10

def choose_difficulty():
  choose = input("Choose a difficulty. Type 'easy' or 'hard': ")
  if choose == "hard":
    return h_guess_num
  else:
    return e_guess_num

def check_if_correct(correct_ans,number,turn):
  if number > correct_ans:
    print("too high")
    return turn -1
  elif number < correct_ans:
    print("too low")
    return turn -1
  else:
    print(f"You got it! The answer was {correct_ans}.")


def game():
  print(logo)
  correct_ans = random.randint(1,100)
  print("Welcome to the Number Guessing Game!")
  print("I'm thinking of a number between 1 and 100.")
  print(f"Pssst, the correct answer is {correct_ans}")

  # 选择跑几次
  turn = choose_difficulty()

  # 因为之前没有number这个var 所以我们创建一个新的
  number = 0
  while number != correct_ans:
    print (f"You have {turn} attempts remaining to guess the number.")
    number = int(input("Make a guess: "))
    ###### 牛逼用法 return
    turn = check_if_correct(correct_ans,number,turn)
    if turn == 0:
      print("You've run out of guesses, you lose.")
      return
    elif number != correct_ans:
      print("Guess again.")

game()
```

```python
# single linked list

class Node(object):
  """节点"""
  def __init__(self, elem): 
    self.elem = elem
    self.next = None

class SingleLinkedList(object):
  """单链表"""
  def __init__(self,node = None):
    self.head = node #私有属性

  def is_empty(self):
    return self.head == None

#########################################
  def length(self):
    """存在几个节点"""
    cur = self.head #第一个节点存在cur
    #count 记录数量
    count = 0 
    while cur != None:
      count += 1
      cur = cur.next
    return count

#########################################   
  def travel(self):
    """遍历整个列表"""
    cur = self.head
    while cur != None:
      print(cur.elem, end = " ")
      cur = cur.next
    print("")

######################################### 
  def add(self,item):
    """链表头部添加元素"""
    node = Node(item)
    node.next = self.head
    self.head = node
    

#########################################
  def append(self,item):
    """链表尾部添加元素"""
    node = Node(item)
    if self.is_empty():
      self.head = node
    else:
      cur = self.head
      while cur.next != None:
        cur = cur.next
      cur.next = node

  #########################################  
  def insert(self,pos,item):
    """指定位置添加元素
       :param pos  从0开始   
    """
    pre = self.head
    count = 0
    if pos <= 0 :
      self.add(item)
    elif pos > self.length() -1: 
      self.append(item)
    else:
      while count < pos - 1:
        count += 1
        pre = pre.next
        # 当循环结束后，pre 等于position - 1
      node = Node(item)
      node.next = pre.next
      pre.next = node

#########################################  
  def remove(self,item):
    """删除节点"""
    cur = self.head
    pre = None
    while cur.next != None:
      if cur.elem == item:
        if cur == self.head:
          self.head = cur.next
        else:
          pre.next = cur.next
        break
          #remove item
      else:
        pre = cur
        cur = cur.next

#########################################
  def search(self,item):
    """查找节点是否存在"""
    cur = self.head
    while cur.next != None:
      if cur.elem == item:
        return True
      else:
        cur = cur.next
    return False
    


if __name__ == "__main__": 
  ll = SingleLinkedList()
  print(ll.is_empty())
  print(ll.length())

  ll.append(1)
  print(ll.is_empty())
  print(ll.length())

  ll.append("A")
  ll.append(3)
  ll.append(4)
  ll.add(8)
  ll.insert(3,9)
  ll.travel()
  #print(ll.search(8))
  ll.remove(9)
  ll.travel()



```
